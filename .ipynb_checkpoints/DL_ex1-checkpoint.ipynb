```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#initialize input data x,target value y and weight parameters w,b
np.random.seed(100)
x = np.linspace(-1,1,100).reshape(100,1)
y = 3*np.power(x,2) + 2 + 0.2*np.random.rand(x.size).reshape(100,1)
w1 = np.random.rand(1,1)
b1 = np.random.rand(1,1)
plt.scatter(x,y)
plt.show()
```

```python
#training model
a = 0.001 #learning rate
for i in range(1,1000):
    y_pred = np.power(x,2)*w1+b1
    d = y_pred - y
    loss = 0.5*np.power(d,2).sum()
    grad_w = np.sum(d*np.power(x,2))
    grad_b = np.sum(d)
    
    w1 -= a*grad_w
    b1 -= a*grad_b    
print(w1,b1)
```

```python
plt.plot(x,y_pred,'r-',label='predict')
plt.scatter(x,y,color='blue',marker='x',label='true') # true data
plt.xlim(-1,1)
plt.ylim(2,6)
plt.legend()
plt.show()
print(w1,b1)
```
